## Metody gradientowe - metoda gradientu prostego

Jednym z najprostszych algorytmów optymalizacji jest metoda gradientu prostego.
Pozukujemy minimum różniczkowalnej funkcji $f:\mathbb{R}^n\rightarrow\mathbb{R}$. Algorytm przedstawia się następująco:
 1. Wybierz punkt startowy $x_0$, $n=0$
 2. Wyznacz nowe $x_{n+1}:=x_n-\alpha\nabla f(x_n)$
 3. $n:=n+1$
 4. Jeżeli $||\nabla f(x_n)||>\epsilon$ i n mniejsze od dopuszczalnej ilości iteracji, to wróć do 2.
Alfa jest krokiem metody. Dla dostatecznie małego alfa $f(x_0)>f(x_1)>f(x_2)>...$
